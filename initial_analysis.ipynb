```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
import ast
```

```python
# In a real scenario, this would be pd.read_csv('your_file.csv', sep='\t')

df = pd.read_csv(r"C:\Users\ether\OneDrive\Documents\i3digitalhealth_assignment\sample_data.csv")

# Clean up column names by removing leading/trailing whitespace
```

```python


# Convert date columns to datetime objects, coercing errors to NaT (Not a Time)
df['start_date'] = pd.to_datetime(df['start_date'], format='%d/%m/%Y', errors='coerce')
df['start_year'] = df['start_date'].dt.year

# --- 2. Define Success and Failure (Operationalization) ---
def define_outcome(status):
    if status == 'COMPLETED':
        return 'Success'
    elif status in ['WITHDRAWN', 'TERMINATED']:
        return 'Failure'
    else:
        return 'Ongoing/Unknown'

df['outcome'] = df['recruitment_status'].apply(define_outcome)

# Create a DataFrame for analysis that only includes trials with a definitive outcome
df_finalized = df[df['outcome'].isin(['Success', 'Failure'])].copy()

# Standardize Phase labels
phase_mapping = {
    'EARLY_PHASE1': 'PHASE1',
    'PHASE1': 'PHASE1',
    'PHASE1/PHASE2': 'PHASE1/2',
    'PHASE2': 'PHASE2',
    'PHASE3': 'PHASE3',
    'PHASE4': 'PHASE4'
}
df_finalized['phase_group'] = df_finalized['phase'].map(phase_mapping).fillna('Other')
df['phase_group'] = df['phase'].map(phase_mapping).fillna('Other') # Also for the main df for stats


# Helper function for plotting
def plot_success_rates(data, title, xlabel, ylabel='Success Rate', min_trials=3):
    """Plots success rates for a given categorical variable."""
    # Filter out categories with too few trials
    category_counts = data.index.get_level_values(0).value_counts()
    valid_categories = category_counts[category_counts >= min_trials].index
    
    if valid_categories.empty:
        print(f"Skipping plot '{title}': No categories met the minimum trial count of {min_trials}.")
        return
        
    plot_data = data[data.index.get_level_values(0).isin(valid_categories)]
    
    # Calculate success rate
    success_rate = (plot_data['Success'] / (plot_data['Success'] + plot_data['Failure'])).unstack().fillna(0)
    success_rate = success_rate.sort_values(by='Success', ascending=False)
    
    plt.figure(figsize=(12, 7))
    sns.barplot(x=success_rate.index, y=success_rate['Success'])
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, 1)
    # Add percentage labels
    for index, value in enumerate(success_rate['Success']):
        plt.text(index, value + 0.02, f'{value:.0%}', ha='center')
    plt.tight_layout()
    plt.show()

print("--- Analysis of Clinical Trial Success Rates ---")

# --- 3. Success Rate by Phase ---
print("\n--- 3.1. Overall Success Rate by Phase ---")
phase_analysis = df_finalized.groupby('phase_group')['outcome'].value_counts().unstack().fillna(0)
phase_analysis['total'] = phase_analysis.sum(axis=1)
phase_analysis['success_rate'] = phase_analysis['Success'] / phase_analysis['total']
print(phase_analysis)

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=phase_analysis.index, y=phase_analysis['success_rate'], palette='viridis')
plt.title('Success Rate by Clinical Trial Phase', fontsize=16)
plt.xlabel('Phase', fontsize=12)
plt.ylabel('Success Rate', fontsize=12)
plt.ylim(0, 1)
for index, value in enumerate(phase_analysis['success_rate']):
    plt.text(index, value + 0.02, f'{value:.0%}', ha='center')
plt.tight_layout()
plt.show()

# --- 4. Success Rate Trends Over Time ---
print("\n--- 4.1. Success Rate Trend by Start Year ---")
time_analysis = df_finalized.groupby(['start_year', 'phase_group'])['outcome'].value_counts().unstack().fillna(0)
time_analysis['total'] = time_analysis.sum(axis=1)
time_analysis['success_rate'] = time_analysis['Success'] / time_analysis['total']
time_analysis_pivot = time_analysis['success_rate'].unstack(level='phase_group')

print("Annual Success Rates by Phase:")
print(time_analysis_pivot)

# Visualization
plt.figure(figsize=(14, 7))
for phase in time_analysis_pivot.columns:
    plt.plot(time_analysis_pivot.index, time_analysis_pivot[phase], marker='o', linestyle='-', label=phase)

plt.title('Trend of Trial Success Rate by Start Year and Phase', fontsize=16)
plt.xlabel('Start Year', fontsize=12)
plt.ylabel('Success Rate', fontsize=12)
plt.legend(title='Phase')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.figtext(0.5, -0.05, "Note: Data for recent years (e.g., post-2024) is highly unreliable due to right-censoring.\nMany trials started in these years are still ongoing and not included in this analysis.", ha="center", fontsize=10, bbox={"facecolor":"orange", "alpha":0.2, "pad":5})
plt.tight_layout()
plt.show()

# --- 5. Data Limitations & Bias Summary ---
print("\n--- 5.1. Data Limitations & Bias ---")
total_trials = len(df)
finalized_trials = len(df_finalized)
ongoing_trials = total_trials - finalized_trials
print(f"Total Trials in Dataset: {total_trials}")
print(f"Trials with Definitive Outcomes (Completed/Withdrawn): {finalized_trials}")
print(f"Ongoing/Unknown Status Trials (Excluded from rates): {ongoing_trials} ({ongoing_trials/total_trials:.1%})")
print("\nKey Caveats for Stakeholders:")
print("1.  **Proxy for Success:** 'Completion' is used as a proxy for success. This does not guarantee clinical efficacy or regulatory approval.")
print("2.  **Right-Censoring:** A significant portion of the dataset consists of ongoing trials. Their exclusion, especially from recent years, biases the trend analysis. Rates for later years are based on very few, short-duration trials.")
print("3.  **Small Sample Sizes:** When segmenting by specific indications, technologies, or targets, the number of trials per group is often too small to draw statistically significant conclusions.")
print("4.  **Dataset Representativeness:** This is a small data extract and may not be representative of the entire oncology landscape, potentially introducing selection bias.")


# --- 6. Deeper Dives: Success by Indication, Technology, and Target ---

# Function to safely evaluate string-formatted lists
def safe_eval(x):
    try:
        # Use ast.literal_eval for safety
        return ast.literal_eval(x)
    except (ValueError, SyntaxError):
        # Return an empty list if it's not a valid list format (e.g., NaN, empty string)
        return []

# Helper function for exploding and analyzing categorical data
def analyze_by_category(df_to_analyze, column_name, plot_title, top_n=10):
    print(f"\n--- Analysis by {column_name} ---")
    
    # Handle nested lists if applicable
    if df_to_analyze[column_name].dtype == 'object' and df_to_analyze[column_name].str.contains(r'\[\[').any():
        df_exploded = df_to_analyze.assign(**{column_name: df_to_analyze[column_name].apply(safe_eval).apply(lambda x: [item for sublist in x for item in sublist])}).explode(column_name)
    else:
        df_exploded = df_to_analyze.assign(**{column_name: df_to_analyze[column_name].apply(safe_eval)}).explode(column_name)

    df_exploded.dropna(subset=[column_name], inplace=True)
    
    # Get top N categories by total number of finalized trials
    top_categories = df_exploded[column_name].value_counts().nlargest(top_n).index
    df_top = df_exploded[df_exploded[column_name].isin(top_categories)]

    analysis = df_top.groupby(column_name)['outcome'].value_counts().unstack().fillna(0)
    analysis['total'] = analysis.sum(axis=1)
    analysis['success_rate'] = analysis['Success'] / analysis['total']
    
    print(analysis.sort_values('total', ascending=False))
    
    # Visualization
    plot_data_for_viz = analysis[['Success', 'Failure']].copy()
    plot_success_rates(plot_data_for_viz.stack().unstack(0), plot_title, xlabel=column_name)

# 6.1 Success Rate by Indication
analyze_by_category(
    df_finalized, 
    'indications', 
    'Success Rate for Top 10 Indications',
    top_n=10
)

# 6.2 Success Rate by Technology
# The 'main_technologies' is a list of lists, requiring flattening.
df_tech = df_finalized.copy()
# Convert the string representation of lists into actual lists
df_tech['main_technologies'] = df_tech['main_technologies'].apply(safe_eval)
# Flatten the list of lists
df_tech['main_technologies'] = df_tech['main_technologies'].apply(lambda x: [item for sublist in x for item in sublist])
# Explode the DataFrame
df_tech_exploded = df_tech.explode('main_technologies')
df_tech_exploded.dropna(subset=['main_technologies'], inplace=True)

# Get top 5 technologies
top_tech = df_tech_exploded['main_technologies'].value_counts().nlargest(5).index
df_top_tech = df_tech_exploded[df_tech_exploded['main_technologies'].isin(top_tech)]
tech_analysis = df_top_tech.groupby('main_technologies')['outcome'].value_counts().unstack().fillna(0)

print("\n--- 6.2. Success Rate by Main Technology ---")
print(tech_analysis)
plot_success_rates(tech_analysis, 'Success Rate for Top 5 Technologies', 'Technology')


# 6.3 Success Rate by Target
analyze_by_category(
    df_finalized, 
    'target_abbreviations', 
    'Success Rate for Top 10 Targets',
    top_n=10
)
```
